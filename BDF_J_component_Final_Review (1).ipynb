```python
# Cell 1: Install Java and set up environment
!apt-get install openjdk-8-jdk-headless -qq > /dev/null
import os
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-8-openjdk-amd64"

# Initialize Spark Session with optimized configurations
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, mean, stddev, corr, when, lit, sum as spark_sum, rank, dense_rank, percent_rank, ntile
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType
from pyspark.ml.feature import VectorAssembler, StandardScaler, PCA, ChiSqSelector
from pyspark.ml.clustering import KMeans, BisectingKMeans, GaussianMixture
from pyspark.ml.classification import RandomForestClassifier, GBTClassifier
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.evaluation import BinaryClassificationEvaluator, RegressionEvaluator
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.sql.window import Window
from pyspark.ml.stat import Correlation, ChiSquareTest
import numpy as np

spark = SparkSession.builder \
    .appName("BigDataAnalysis") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .config("spark.sql.shuffle.partitions", "4") \
    .getOrCreate()

print("Spark session created successfully!")
```

```python
# Cell 2: Load data with schema definition and partitioning
schema = StructType([
    StructField("Age", IntegerType(), True),
    StructField("Attrition", StringType(), True),
    StructField("BusinessTravel", StringType(), True),
    StructField("DailyRate", IntegerType(), True),
    StructField("Department", StringType(), True),
    StructField("DistanceFromHome", IntegerType(), True),
    StructField("Education", IntegerType(), True),
    StructField("EducationField", StringType(), True),
    StructField("EmployeeCount", IntegerType(), True),
    StructField("EmployeeNumber", IntegerType(), True),
    StructField("EnvironmentSatisfaction", IntegerType(), True),
    StructField("Gender", StringType(), True),
    StructField("HourlyRate", IntegerType(), True),
    StructField("JobInvolvement", IntegerType(), True),
    StructField("JobLevel", IntegerType(), True),
    StructField("JobRole", StringType(), True),
    StructField("JobSatisfaction", IntegerType(), True),
    StructField("MaritalStatus", StringType(), True),
    StructField("MonthlyIncome", IntegerType(), True),
    StructField("MonthlyRate", IntegerType(), True),
    StructField("NumCompaniesWorked", IntegerType(), True),
    StructField("Over18", StringType(), True),
    StructField("OverTime", StringType(), True),
    StructField("PercentSalaryHike", IntegerType(), True),
    StructField("PerformanceRating", IntegerType(), True),
    StructField("RelationshipSatisfaction", IntegerType(), True),
    StructField("StandardHours", IntegerType(), True),
    StructField("StockOptionLevel", IntegerType(), True),
    StructField("TotalWorkingYears", IntegerType(), True),
    StructField("TrainingTimesLastYear", IntegerType(), True),
    StructField("WorkLifeBalance", IntegerType(), True),
    StructField("YearsAtCompany", IntegerType(), True),
    StructField("YearsInCurrentRole", IntegerType(), True),
    StructField("YearsSinceLastPromotion", IntegerType(), True),
    StructField("YearsWithCurrManager", IntegerType(), True)
])

# Upload the file if not already present
from google.colab import files
uploaded = files.upload()

# Load the data
df = spark.read.csv("watson.csv", header=True, schema=schema).repartition(4, "Department")
df.persist()

# Cell 2: Add these lines to show data info
print(f"Total records: {df.count()}")
df.show(5)
```

```python
# Cell 3: Advanced Feature Engineering using distributed operations
df_enhanced = df.withColumn("IncomePerHour", col("MonthlyIncome") / 160) \
               .withColumn("IncomePerYear", col("MonthlyIncome") * 12 / (col("YearsAtCompany") + 1)) \
               .withColumn("PromoRate", col("YearsAtCompany") / (col("YearsSinceLastPromotion") + 1)) \
               .withColumn("StabilityRatio", col("YearsAtCompany") / col("Age")) \
               .withColumn("CareerProgression", col("JobLevel") / (col("YearsAtCompany") + 1)) \
               .withColumn("SatisfactionIndex", (col("JobSatisfaction") + col("EnvironmentSatisfaction") +
                                                 col("RelationshipSatisfaction") + col("WorkLifeBalance")) / 4) \
               .withColumn("IncomeSatisfactionRatio", col("MonthlyIncome") / (col("JobSatisfaction") + 1))
df_enhanced.show(5)
```

```python
# Cell 4: Window Functions for Employee Ranking
window_spec = Window.partitionBy("Department").orderBy(col("MonthlyIncome").desc())
window_spec2 = Window.partitionBy("JobRole").orderBy(col("YearsAtCompany").desc())

df_ranked = df_enhanced.withColumn("DeptIncomeRank", rank().over(window_spec)) \
                     .withColumn("DeptIncomePercentile", percent_rank().over(window_spec)) \
                     .withColumn("JobTenureRank", dense_rank().over(window_spec2)) \
                     .withColumn("JobTenile", ntile(4).over(window_spec2))
df_ranked.show(5)
```

```python
# Cell 5: Multi-level Aggregations with Complex Metrics
dept_job_stats = df_ranked.groupBy("Department", "JobRole") \
                         .agg(
                             count("*").alias("EmployeeCount"),
                             mean("MonthlyIncome").alias("AvgIncome"),
                             stddev("MonthlyIncome").alias("StdDevIncome"),
                             mean("YearsAtCompany").alias("AvgTenure"),
                             mean("PromoRate").alias("AvgPromoRate"),
                             mean("SatisfactionIndex").alias("AvgSatisfaction"),
                             spark_sum(when(col("Attrition") == "Yes", 1).otherwise(0)).alias("AttritionCount")
                         ) \
                         .withColumn("AttritionRate", col("AttritionCount") / col("EmployeeCount")) \
                         .orderBy(col("Department"), col("AttritionRate").desc())
dept_job_stats.show()
```

```python
# Cell 6: Correlation Matrix Analysis
numerical_cols = ["Age", "MonthlyIncome", "YearsAtCompany", "YearsInCurrentRole",
                 "JobSatisfaction", "PerformanceRating", "WorkLifeBalance", "SatisfactionIndex"]

assembler = VectorAssembler(inputCols=numerical_cols, outputCol="features")
df_vector = assembler.transform(df_enhanced.select(numerical_cols))

matrix = Correlation.corr(df_vector, "features").collect()[0][0]
corr_matrix = matrix.toArray().tolist()
print("Correlation Matrix:")
for i, row in enumerate(corr_matrix):
    print(f"{numerical_cols[i]}: {row}")
```

```python
# Cell 7: Chi-Square Test for Categorical Variables
from pyspark.ml.feature import StringIndexer

# Convert Attrition to numerical label first
label_indexer = StringIndexer(inputCol="Attrition", outputCol="label")
df_label = label_indexer.fit(df_enhanced).transform(df_enhanced)

categorical_cols = ["Department", "EducationField", "Gender", "JobRole", "MaritalStatus", "OverTime"]
chi_results = []

for cat_col in categorical_cols:
    # Convert categorical column to numerical index
    indexer = StringIndexer(inputCol=cat_col, outputCol=cat_col+"_index")
    df_indexed = indexer.fit(df_label).transform(df_label)

    # Create vector for the single feature
    assembler = VectorAssembler(inputCols=[cat_col+"_index"], outputCol="features")
    df_vector = assembler.transform(df_indexed)

    # Run Chi-Square test
    chi_result = ChiSquareTest.test(df_vector, "features", "label").head()
    # Convert numpy float64 to Python float
    p_value = float(chi_result.pValues[0])
    chi_results.append((cat_col, p_value, chi_result.degreesOfFreedom[0]))

# Create DataFrame with explicit schema
from pyspark.sql.types import StructType, StructField, StringType, FloatType, IntegerType
schema = StructType([
    StructField("Feature", StringType(), True),
    StructField("PValue", FloatType(), True),
    StructField("DegreesOfFreedom", IntegerType(), True)
])
chi_df = spark.createDataFrame(chi_results, schema)

# Cell 7: Add this line to show chi-square results
chi_df.show()
```

```python
# Cell 8: Advanced Clustering - Employee Segmentation
cluster_features = ["Age", "MonthlyIncome", "YearsAtCompany", "JobSatisfaction",
                   "WorkLifeBalance", "PromoRate", "SatisfactionIndex"]

assembler = VectorAssembler(inputCols=cluster_features, outputCol="features")
df_cluster = assembler.transform(df_enhanced)

scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures")
scaler_model = scaler.fit(df_cluster)
df_cluster = scaler_model.transform(df_cluster)

# K-Means Clustering
kmeans = KMeans(k=5, seed=1, featuresCol="scaledFeatures", predictionCol="kmeans_cluster")
kmeans_model = kmeans.fit(df_cluster)
kmeans_predictions = kmeans_model.transform(df_cluster)

# Cell 8: Add this line to show cluster predictions
kmeans_predictions.select("kmeans_cluster").show(5)
```

```python
# Cell 9: Bisecting K-Means Clustering
bkm = BisectingKMeans(k=5, seed=1, featuresCol="scaledFeatures", predictionCol="bkm_cluster")
bkm_model = bkm.fit(df_cluster)
bkm_predictions = bkm_model.transform(df_cluster)
# Cell 9: Add this line to show BKM predictions
bkm_predictions.select("bkm_cluster").show(5)
```

```python
# Cell 10: Gaussian Mixture Model Clustering
gmm = GaussianMixture(k=5, seed=1, featuresCol="scaledFeatures", predictionCol="gmm_cluster")
gmm_model = gmm.fit(df_cluster)
gmm_predictions = gmm_model.transform(df_cluster)
# Cell 10: Add this line to show GMM predictions
gmm_predictions.select("gmm_cluster").show(5)
```

```python
# Cell 11: Advanced Classification - Attrition Prediction (Corrected)
from pyspark.ml.feature import StringIndexer, Imputer
from pyspark.sql.functions import col, lit

# Convert Attrition to numerical label manually to ensure binary encoding
df_class = df_enhanced.withColumn("label", when(col("Attrition") == "Yes", 1).otherwise(0))

feature_cols = ["Age", "DistanceFromHome", "Education", "JobLevel", "MonthlyIncome",
               "NumCompaniesWorked", "TotalWorkingYears", "TrainingTimesLastYear",
               "YearsAtCompany", "YearsSinceLastPromotion", "IncomePerHour",
               "PromoRate", "StabilityRatio", "SatisfactionIndex"]

# Check for columns that are entirely null and fill them with 0
for col_name in feature_cols:
    # Check if all values in the column are null
    if df_class.filter(col(col_name).isNull()).count() == df_class.count():
        df_class = df_class.withColumn(col_name, lit(0))

# Handle remaining null values with imputation
imputer = Imputer(inputCols=feature_cols, outputCols=feature_cols, strategy="median")
df_class = imputer.fit(df_class).transform(df_class)

# Create feature vector
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features", handleInvalid="skip")
df_class = assembler.transform(df_class)

# Verify the label has only two distinct values
distinct_labels = df_class.select("label").distinct().collect()
print(f"Distinct labels: {[row['label'] for row in distinct_labels]}")

train_data, test_data = df_class.randomSplit([0.8, 0.2], seed=42)

# Cell 11: Add these lines to show prepared data
print(f"Training data count: {train_data.count()}")
print(f"Test data count: {test_data.count()}")
```

```python
# Cell 13: Gradient Boosted Trees Classification (Corrected)
gbt = GBTClassifier(featuresCol="features", labelCol="label", seed=42, maxIter=10)
gbt_model = gbt.fit(train_data)
gbt_predictions = gbt_model.transform(test_data)
# Cell 13: Add these lines to show GBT predictions
print("Gradient Boosted Trees Predictions:")
gbt_predictions.select("label", "prediction", "probability").show(5)
```

```python
# Cell 14: Advanced Regression - Income Prediction (Corrected)
# Create a separate feature list for regression to avoid conflicts
reg_feature_cols = ["Age", "DistanceFromHome", "Education", "JobLevel",
                   "NumCompaniesWorked", "TotalWorkingYears", "TrainingTimesLastYear",
                   "YearsAtCompany", "YearsSinceLastPromotion", "IncomePerHour",
                   "PromoRate", "StabilityRatio", "SatisfactionIndex"]

df_reg = df_enhanced.select(reg_feature_cols + ["MonthlyIncome"])

# Check for columns that are entirely null and fill them with 0
for col_name in reg_feature_cols + ["MonthlyIncome"]:
    # Check if all values in the column are null
    if df_reg.filter(col(col_name).isNull()).count() == df_reg.count():
        df_reg = df_reg.withColumn(col_name, lit(0))

# Handle remaining null values with imputation
imputer = Imputer(inputCols=reg_feature_cols + ["MonthlyIncome"], outputCols=reg_feature_cols + ["MonthlyIncome"], strategy="median")
df_reg = imputer.fit(df_reg).transform(df_reg)

assembler = VectorAssembler(inputCols=reg_feature_cols, outputCol="features", handleInvalid="skip")
df_reg = assembler.transform(df_reg).withColumnRenamed("MonthlyIncome", "target")

train_data, test_data = df_reg.randomSplit([0.8, 0.2], seed=42)

rf_reg = RandomForestRegressor(featuresCol="features", labelCol="target", seed=42)
rf_reg_model = rf_reg.fit(train_data)
rf_reg_predictions = rf_reg_model.transform(test_data)
# Cell 14: Add these lines to show regression results
print("Random Forest Regression Results:")
rf_reg_predictions.select("target", "prediction").show(5)
```

```python
# Cell 15: Feature Importance Analysis (Corrected)
importances = rf_reg_model.featureImportances.toArray()
feature_importance_list = [(reg_feature_cols[i], float(importances[i])) for i in range(len(reg_feature_cols))]
feature_importance_df = spark.createDataFrame(feature_importance_list, ["Feature", "Importance"]).orderBy(col("Importance").desc())
# Cell 15: Add this line to show feature importance
feature_importance_df.show()
```

```python
# Cell 18: Advanced Time Series Analysis - Tenure Patterns
df_tenure = df_enhanced.withColumn("TenureCohort",
                                  when(col("YearsAtCompany") < 2, "0-2 years")
                                  .when((col("YearsAtCompany") >= 2) & (col("YearsAtCompany") < 5), "2-5 years")
                                  .when((col("YearsAtCompany") >= 5) & (col("YearsAtCompany") < 10), "5-10 years")
                                  .otherwise("10+ years"))

tenure_income = df_tenure.groupBy("TenureCohort") \
                        .agg(
                            count("*").alias("EmployeeCount"),
                            mean("MonthlyIncome").alias("AvgIncome"),
                            mean("IncomePerYear").alias("IncomeGrowth"),
                            mean("PromoRate").alias("PromotionRate"),
                            spark_sum(when(col("Attrition") == "Yes", 1).otherwise(0)).alias("AttritionCount")
                        ) \
                        .withColumn("AttritionRate", col("AttritionCount") / col("EmployeeCount")) \
                        .orderBy(col("TenureCohort"))


# Cell 18: Add this line to show tenure patterns
tenure_income.show()
```

```python
# Cell 19: Advanced Risk Scoring with Weighted Metrics
risk_factors = {
    "DistanceFromHome": 0.15,
    "WorkLifeBalance": 0.25,
    "JobSatisfaction": 0.20,
    "OverTime": 0.20,
    "YearsSinceLastPromotion": 0.10,
    "RelationshipSatisfaction": 0.10
}

df_risk = df_enhanced.withColumn("RiskScore",
                                (col("DistanceFromHome") * risk_factors["DistanceFromHome"] +
                                 (5 - col("WorkLifeBalance")) * risk_factors["WorkLifeBalance"] +
                                 (5 - col("JobSatisfaction")) * risk_factors["JobSatisfaction"] +
                                 when(col("OverTime") == "Yes", 1).otherwise(0) * risk_factors["OverTime"] +
                                 col("YearsSinceLastPromotion") * risk_factors["YearsSinceLastPromotion"] +
                                 (5 - col("RelationshipSatisfaction")) * risk_factors["RelationshipSatisfaction"]))

high_risk = df_risk.filter(col("RiskScore") > 3.0) \
                 .select("EmployeeNumber", "Department", "JobRole", "RiskScore", "Attrition") \
                 .orderBy(col("RiskScore").desc())

# Cell 19: Add this line to show high-risk employees
high_risk.show(10)
```

```python
# Cell 20: Advanced Anomaly Detection - Income Outliers
income_stats = df_enhanced.groupBy("JobRole") \
                         .agg(
                             mean("MonthlyIncome").alias("mean_income"),
                             stddev("MonthlyIncome").alias("stddev_income")
                         )

df_with_stats = df_enhanced.join(income_stats, "JobRole", "left") \
                          .withColumn("income_zscore",
                                     (col("MonthlyIncome") - col("mean_income")) / col("stddev_income"))

income_outliers = df_with_stats.filter((col("income_zscore") > 3) | (col("income_zscore") < -3)) \
                              .select("EmployeeNumber", "JobRole", "MonthlyIncome", "income_zscore") \
                              .orderBy(col("income_zscore").desc())
# Cell 20: Add this line to show income outliers
income_outliers.show(10)
```
